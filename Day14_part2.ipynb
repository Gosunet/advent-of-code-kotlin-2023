```kotlin
fun List<String>.transpose(): List<String> {
    val maxLength = this.maxByOrNull { it.length }?.length ?: 0
    return List(maxLength) { index ->
        this.map { str -> if (index < str.length) str[index] else ' ' }.joinToString("")
    }
}
```

```kotlin
fun List<String>.rotate90DegreesClockwise(): List<String> {
    val maxLength = this.maxByOrNull { it.length }?.length ?: 0

    return List(maxLength) { index ->
        this.reversed().map { str ->
            if (index < str.length) str[index] else ' '
        }.joinToString("")
    }
}
```

```kotlin
fun List<String>.rotateDegreesClockwise(rotations: Int): List<String> {
    var rotatedList = this
    repeat(rotations % 4) {
        rotatedList = rotatedList.rotate90DegreesClockwise()
    }
    return rotatedList
}
fun List<String>.rotate90DegreesCounterClockwise(): List<String> {
    val maxLength = this.maxByOrNull { it.length }?.length ?: 0

    return List(maxLength) { index ->
        this.map { str ->
            val reversedIndex = maxLength - 1 - index
            if (reversedIndex < str.length) str[reversedIndex] else ' '
        }.joinToString("")
    }
}
```

```kotlin
fun List<String>.roll(reversed: Boolean = false) = this.map { line ->
  line.split("(?=#)".toRegex()).map {
    it.toCharArray().sortedWith(
      compareBy {
        when (it) {
          '#' -> 0
          'O' -> if (!reversed) 1 else 2
          else -> if (!reversed) 2 else 1
        }
      }
    ).joinToString("")
  }.joinToString("")
}
```
